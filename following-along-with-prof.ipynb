### Plotting with matplotlib!

```python
# Importing numpy and matplotlib -
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining the potential energy function -
def potential_energy(x, k=1.0):
    return 0.5 * k * x**2         # k is the spring constant!

# Testing fuction -
potential_energy(2, 10)
```

```python
# Defining mesh for plotting -
x = np.linspace(-10, 10, 1000)        # Syntax is (x_min, x_max, number of points)

# Defining y values from x list -
y = potential_energy(x)

# Testing output -
print(np.transpose([x, y]))
```

```python
# Plotting the function -
plt.figure(figsize=(3,3))       # Figure size in inches
plt.plot(x, y, color="green", linewidth=0.5)

# Adding axes labels -
plt.xlabel("Displacement (arbitrary units)")
plt.ylabel("Potential Energy (arbitrary units)")

# Changing limits of x and y -
plt.xlim([-4,4])
plt.ylim([-1,10])

plt.show()          # Combines all elements of the plot!
```

### Contour plots

```python
# Importing numpy and matplotlib -
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining the potential energy function -
def potential_energy_2D(x, y, kx=1.0, ky=1.0):
    return 0.5 * (kx * x**2 + ky * y**2)          # (kx, ky) are the spring constants!

# Testing fuction -
potential_energy_2D(2, 1)
```

```python
# Defining mesh for plotting -
x = np.linspace(-5, 5, 100)        # Syntax is (x_min, x_max, number of points)
y = np.linspace(-5, 5, 100)        # Syntax is (y_min, y_max, number of points)

# Combining 1D meshes into meshgrid -
X, Y = np.meshgrid(x,y)

# Defining Z values -
Z = potential_energy_2D(X,Y)
# Testing the mesh -
print(X)
print()
print(Y)
print()
print(Z)
```

```python
# Constructing the contour plot -
plt.figure(figsize=(3,3))       # Figure size in inches
plt.contourf(X, Y, Z)

# Adding a colorbar legend -
plt.colorbar(label="Potential Energy")
# Adding axes labels -
plt.xlabel("Displacement (arbitrary units)")
plt.ylabel("Potential Energy (arbitrary units)")

# Changing limits of x and y -
plt.xlim([-5,5])
plt.ylim([-5,5])

plt.show()          # Combines all elements of the plot!
```
