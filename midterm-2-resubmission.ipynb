### Problem 1: Quartic Anharmonic Oscillator with Positive Quadratic Term

Consider the anharmonic oscillator described by the potential energy function:
$$
U(Q) = \frac{1}{2} K Q^2 + \frac{1}{4} D Q^4
$$

1. **Harmonic Frequency**
   * (a) Assuming small oscillations, derive an expression for the harmonic frequency of this system when $Q$ is close to zero.

2. **Comparison of Terms in the Potential**
   * (a) Find an expression for the value of $Q$ at which the fourth-order term $\frac{1}{4} D Q^4$ becomes larger than the second-order term $\frac{1}{2} K Q^2$. Discuss what this indicates about the relative importance of each term at larger amplitudes.

3. **Equation of Motion**
   * (a) Write down the equation of motion for this system.

4. **Amplitude-Dependent Frequency (Computational)**
   * (a) Numerically solve for the angular frequency of the oscillator as a function of initial amplitude $Q_0$ (for zero initial velocity). One potential approach is to solve the equation of motion using a differential equation solver and then apply a root-finding algorithm to locate the period.
   * (b) Plot the frequency as a function of the starting amplitude $Q_0$.
   * (c) Describe and explain how the frequency changes with amplitude.


#### 1. Harmonic Frequency

As Q approaches zero (when Q < 1) in the potential energy equation
$$U(Q) = \frac{1}{2} K Q^2 + \frac{1}{4} D Q^4$$
the fourth-order anharmonic term $\frac{1}{4} D Q^4$
becomes negligible (because as Q approaches zero, the fourth-order term with Q to the fourth power approaches zero much more rapidly than the second-order term with Q to the second power), leaving just the equation for a simple harmonic oscillator
$U(Q) = \frac{1}{2} K Q^2$
for finding the harmonic frequency. The frequency of a simple harmonic oscillator is given as
$$\omega = \sqrt{\frac{K}{m}}$$
where K is the spring constant of the oscillator, and m is the mass of the object.


#### 2. Comparison of Terms in the Potential

To find an expression where the value of Q causes the fourth-order term to become greater than the second-order term, we would set up an inequality between the terms such as

$$\frac{1}{2} K Q^2 < \frac{1}{4} D Q^4$$

and then solve for Q -

$$2 K Q^2 < D Q^4$$ 
$$2 K < D Q^2$$
$$\frac{2K}{D} < Q^2$$
$$\sqrt{\frac{2K}{D}} < Q$$

Therefore, the values of Q at which the fourth-order term becomes greater than the second-order term are

$$Q > \sqrt{\frac{2K}{D}}$$

where Q is the displacement of the oscillator from its equilibrium position, K is the spring constant of the oscillator, and D is the strength of the anharmonicity of the oscillator. Given that the second-order term $\frac{1}{2} K Q^2$ is the expression for the portential energy of a simple harmonic oscillator, and the fourth-order term $\frac{1}{4} D Q^4$ is the term that accounts for the anharmonic contribution to the potential energy of an anharmonic oscillator, the derived expression $Q > \sqrt{\frac{2K}{D}}$ represents the displacement values at which the fourth-order anharmonic term becomes more significant than the second-order term. For amplitudes at values of Q smaller than this minimum value, the harmonic term is more significant, meaning the system will behave as simple harmonic motion. For amplitudes at values of Q at or larger than this minumim value, the anharmonic term is more significant and starts to influence the system's behavior, so the system will no longer behave as simple harmonic motion.


#### 3. Equation of Motion

The equation for the potential energy of this system is given as

$$U(Q) = \frac{1}{2} K Q^2 + \frac{1}{4} D Q^4$$

To find the equation of motion for this system, first we would need to find the expression for the force of the system. The relationship between potential energy and force is the force is the negative of the potential energy U with respect to Q, the displacement of the object in the system. The derivative of U(Q) is found by differentiating both the second-order and the fourth-order terms -

$$\frac{d}{dQ} (\frac{1}{2} K Q^2) = K Q$$
$$\frac{d}{dQ} (\frac{1}{4} D Q^4) = D Q^3$$

Combined and multiplying by -1, the derivative of the potential energy U with respect to Q is

$$F(Q) = -\frac{dU}{dQ} = -K Q - D Q^3$$

With this expression of force in the anharmonic oscillator, we can then find the equation of motion using Newton's Second Law, force equals mass times acceleration. Knowing that displacement is the second derivative of displacement with respect to time, we can rewrite the acceleration in the system as

$$F(Q) = m \frac{d^2 Q}{dt^2}$$

Then, with these two expressions for the force of the system, we can set them equal to each other, giving us the equation of motion of this system -

$$m \frac{d^2 Q}{dt^2} = -K Q - D Q^3$$
$$\frac{d^2 Q}{dt^2} + \frac{K Q}{m} + \frac{D Q^3}{m} = 0$$


#### 4. Amplitude-Dependent Frequency

Solving for the angular frequency of the oscillator as a function of its initial amplitude and plotting -


```python
# Importing necessary libraries -
import numpy as np
import matplotlib.pyplot as plt               # For plotting frequency vs. initial amplitude
from scipy.integrate import solve_ivp         # For solving for angular frequency

# Defining constants -
K = 1.0       # Spring constant
D = 1.0       # Anharmonicity strength constant
m = 1.0       # Mass of oscillating object

# Defining potential energy function -
def U(Q):
    return (1/2) * K * Q**2 + (1/4) * D * Q**4

# Defining force derived from potential energy -
def force(Q):
    return -K * Q - D * Q**3

# Defining system of differential equations for the equation of motion -
def oscillator(t, y):
    Q, v = y                                      # Rewriting d^2Q/dt^2 in terms of velocity
    return [v, force(Q) / m]                      # Using Newton's Second Law

# Defining function to find period for given position -
def calculate_period(Q0):
    initial_conditions = [Q0, 0]                  # Initial conditions position and velocity
    t_span = (0, 100)                             # Time span
    t_eval = np.linspace(0, 100, 200)             # Points to evaluate solution for period

    # Function for solving the system of first order differential equations -
    sol = solve_ivp(oscillator, t_span, initial_conditions, t_eval=t_eval)
    
    # Finding time it takes for oscillator to return to initial position Q0 -
    times = sol.t
    positions = sol.y[0]

    # Finding period by finding when oscillator returns to initial position -
    crossings = np.where(np.diff(np.sign(positions - Q0)))[0]

    if len(crossings) >= 2:
        period = times[crossings[1]] - times[crossings[0]]
        return period
    else:
        print(f"Not enough crossings detected to find period")
        return np.nan

# Testing amplitudes -
Q0_values = np.linspace(0.1, 2.0, 200)           # Testing amplitudes from 0.1 to 2.0
frequencies = []                                 # Resulting frequensies

# Solving for frequencies -
for Q0 in Q0_values:
    period = calculate_period(Q0)
    frequency = 1 / period                      # From period - frequency relationship
    frequencies.append(frequency)

# Plotting results -
plt.figure(figsize=(10, 6))
plt.plot(Q0_values, frequencies)
plt.title('Frequency vs. Initial Amplitude for Anharmonic Oscillator')
plt.xlabel('Initial Amplitude $Q_0$')
plt.ylabel('Frequency (Hz)')
plt.grid()
plt.show()

print(calculate_period(Q0))
print(frequency)
```

The plot above visualized what was found in the previous problems- that as the initial displacement and initial amplitude of the anharmonic system increases, the contribution of the fourth-order term in the potential energy equation becomes more significant. From the equation

$$F(Q) = -\frac{dU}{dQ} = -K Q - D Q^3$$

we see that as the fourth-order term increases, the restoring force of the oscillator weakens, which in turn would decrease the frequency of the oscillator! The plot shown does not have a constant slope, because of the anharmonicity of the system- the quartic term causes steep increases in the potential energy.


### Problem 2: Double-Well Potential with Negative Quadratic Term

#### 1. Equilibrium Points and Potential Energy

The equilibrium points for the anharmonic oscillator expression will occur when the force acting on the oscillator is zero. As seen in the previous problems, the relationship between the force and the potential energy of the oscillator is given by 

$$F(Q) = -\frac{dU}{dQ} = K Q - D Q^3$$

(adjusted for the negative quadratic term). Solving this expression for when $Q = 0$ gives us

$$KQ - DQ^3 = 0$$

$$Q(K - DQ^2) = 0$$

$$K - DQ^2 = 0 \implies DQ^2 = K \implies Q^2 = \frac{K}{D}$$

So, the equilibrium points for this anharmonic oscillator are

$$Q = \pm \sqrt{\frac{K}{D}}$$

$$Q = 0$$

At $Q = 0$, $U(Q)$ is of course

$$U(0) = -\frac{1}{2} K (0)^2 + \frac{1}{4} D (0)^4 = 0$$

At the other equilibrium points, U(Q) is calculated as

$$U(\sqrt{\frac{K}{D}}) = -\frac{1}{2} K (\sqrt{\frac{K}{D}})^2 + \frac{1}{4} D (\sqrt{\frac{K}{D}})^4$$

$$U(\sqrt{\frac{K}{D}}) = -\frac{1}{2} K ({\frac{K}{D}}) + \frac{1}{4} D ({\frac{K^2}{D^2}})$$

$$U(\sqrt{\frac{K}{D}}) = -\frac{K^2}{2D} + \frac{K^2}{4D} = -\frac{2K^2}{4D} + \frac{K^2}{4D} = -\frac{K^2}{4D}$$

Since both terms are even-powered, this solution is the same for both the positive and negative Q terms. The shape of this potential energy curve would be double-well shaped, where the local minimums are at $Q = \pm \sqrt{\frac{K}{D}}$ and the local maximum is at $Q = 0$. The curve will steeply increase as Q moves away from the local minimum points betcause of the quartic anharmonic term.


#### 2. Frequency Near Equilibrium

To find the frequency of oscillations around the local minimums of the anharmonic oscillation potential energy expression, we need to take the second derivative of the potential energy expression with respect to the displacement Q. We already found that the first derivative of the equation (the force) is given by

$$F(Q) = \frac{dU}{dQ} = -K Q + D Q^3$$

The second derivative of the equation is given by

$$\frac{d^2U}{dQ^2} = -K + 3D Q^2$$

Evaluating this at the positive Q equilibrium solution, $\sqrt{\frac{K}{D}}$, gives

$$-K + 3D(\sqrt{\frac{K}{D}})^2 = 2K$$

Finally, substituting this into the formula for frequency of a harmonic oscillator is

$$\omega = \sqrt{\frac{2K}{m}}$$


#### 3. Amplitude-Dependent Frequency

```python
# Importing necessary libraries -
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.optimize import root_scalar

# Defining constants -
K = 1.0       # Spring constant
D = 1.0       # Anharmonicity strength constant
m = 1.0       # Mass of oscillating object

# Defining potential energy function -
def U(Q):
    return -(1/2) * K * Q**2 + (1/4) * D * Q**4

# Defining force derived from potential energy -
def force(Q):
    return K * Q - D * Q**3

# Defining system of differential equations for the equation of motion -
def oscillator(t, y):
    Q, v = y                                      # Rewriting d^2Q/dt^2 in terms of velocity
    return [v, force(Q) / m]                      # Using Newton's Second Law

# Defining function to find the period for given position -
def calculate_period(Q0):
    initial_conditions = [Q0, 0.0]                # Initial conditions position and velocity
    t_span = (0, 100)                             # Time span
    t_eval = np.linspace(0, 100, 1000)            # Points to evaluate solution for period

    # Function for solving the system of first order differential equations -
    sol = solve_ivp(oscillator, t_span, initial_conditions, t_eval=t_eval)
    
    # Finding time it takes for oscillator to return to initial position Q0 -
    times = sol.t
    positions = sol.y[0]

    # Finding period by finding when oscillator returns to initial position -
    crossings = np.where(np.diff(np.sign(positions - Q0)))[0]
    period = times[crossings[1]] - times[crossings[0]]

    return period

# Testing amplitudes -
Q0_values = np.linspace(-2, 2, 50)           # Testing amplitudes
frequencies = []                             # Resulting frequensies

# Solving for frequencies -
for Q0 in Q0_values:
    period = calculate_period(Q0)
    frequency = 1 / period                      # From period - frequency relationship
    frequencies.append(frequency)

# Plotting results -
plt.figure(figsize=(10, 6))
plt.plot(Q0_values, frequencies)
plt.title('Frequency vs. Initial Amplitude for Anharmonic Oscillator')
plt.xlabel('Initial Amplitude $Q_0$')
plt.ylabel('Frequency (Hz)')
plt.grid()
plt.show()

print(calculate_period(Q0))
print(frequency)
```

In the plot above, we can see that the frequency of the system peaks at the local minimums of the potential energy expression and reaches a valley at the local maximum. The frequency increases sharply near (but not exactly at) $ Q_0 = \pm \sqrt{2}$ and reaches a minimum near $Q = 0$. When the system is at equilibrium and the force acting upon the oscillator is zero, in the case of the $ Q_0 = \pm \sqrt{2}$ equilibrium points, the frequency of the anharmonic motion reaches its maximum. The relationship between initial amplitude and frequency is not exactly what I expected from the derived equations in previous parts- the maximums and minimum are not precisely at the equilibrium points.

#### 4. Discussion of Motion Across Regimes

a. We observe in the plot above that when the initial amplitude Q is near -1 and 1, the frequency of the oscillations reach their peak, but there are also smaller pears near -1.75 and 1.75. As the initial amplitudes become larger, the object is able to cross between the central barrier of the wells, so the potential energy of the object had to be great enough to overcome the potential barrier- which means the quartic term is contributing significantly in the equation for potential energy. As seen before, as the quartic anharmonic term dominates in the equation for potential energy, the system behaves less and less like simple harmonic motion. For systems where the motion is confined within a single well, meaning simple harmonic systems, the restoring force is not affected- this is why there are significantly larger peaks around $Q = \pm \sqrt{\frac{K}{D}}$.

b. Anharmonic terms in equations of oscillating motion significantly affect the shape of potential energy distributions, because the anharmonicity greatly affects the restoring force of the system! As the initial amplitude increases (and in turn, the frequency decreases), the oscillator behaves less and less like simple harmonic motion due to the decrease in restoring force (because of the quartic term).
